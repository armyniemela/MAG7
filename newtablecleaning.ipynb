```python
# this I asked about from ChatGPT, couldn't do it my self!!
# Get the values from the owner columns into a single series
owner_columns = ['Holders AAPL', 'HoldersMETA', 'HoldersAMZN', 'HoldersMSFT', 'HoldersGOOG', 'HoldersTSLA', 'HoldersNVDA']
all_owners = pd.concat([mag_df[col] 
for col in owner_columns], ignore_index=True)
unique_owners = all_owners.unique()
unique_owners
```

```python
# to get a better look:

for owner in unique_owners:
    print(owner)
    print('-----')
```
